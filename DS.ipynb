```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('paddy_prod.csv')
```

```python
print(df)
```

Removing columns with vlues with Acres and Bushels and keeping Hectare and KG units since those are the metric units

```python
columns_to_remove = ['Sown 000 Acres', 'Harvested 000 Acres', 'Average Yield Bushels/Acre', 'Production 000 Bushels']
df = df.drop(columns=columns_to_remove)
```

```python
print(df)
```

```python
x = df['Yala Season']
y = df['Sown 000 Ha']

plt.plot(x, y, marker='o', linestyle='-')

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label="Trend Line")

plt.xlabel('Year')
plt.ylabel('Sown 000 Ha')
plt.title('Timeline')

plt.grid(True)

plt.show()
```

```python
x = df['Yala Season']
y1 = df['Sown 000 Ha']
y2 = df['Harvested 000 Ha.']
y3 = df['Average Yield Kg/Ha']

fig, ax1 = plt.subplots()

# ax1.plot(x, y1, color='tab:blue', marker='o', linestyle='-', label='Line 1')
ax1.plot(x, y2, color='tab:red', marker='o', linestyle='-', label='Line 2')

ax1.set_xlabel('Year')
ax1.set_ylabel('Sown 000 Ha', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()

ax2.plot(x, y3, color='tab:green', marker='s', linestyle='--', label='Line 3')

ax2.set_ylabel('Average Yield Kg/Ha', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

ax1.grid(True)
plt.title('Dual Timeline')
plt.show()
```

```python
x = df['Yala Season']
y1 = df['Sown 000 Ha']
y2 = df['Harvested 000 Ha.']
y3 = df['Average Yield Kg/Ha']

fig, ax1 = plt.subplots()

# ax1.plot(x, y1, color='tab:blue', marker='o', linestyle='-', label='Line 1')
# ax1.plot(x, y2, color='tab:red', marker='o', linestyle='-', label='Line 2')

z1 = np.polyfit(x, y2, 1)
p1 = np.poly1d(z1)
ax1.plot(x, p1(x), "r--", label="Trend Line 1")

ax1.set_xlabel('Year')
ax1.set_ylabel('Sown 000 Ha', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()

# ax2.plot(x, y3, color='tab:green', marker='s', linestyle='--', label='Line 3')

# Trend line
z2 = np.polyfit(x, y3, 1)
p2 = np.poly1d(z2)
ax2.plot(x, p2(x), "g--", label="Trend Line 2")

ax2.set_ylabel('Average Yield Kg/Ha', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

ax1.grid(True)

plt.title('Dual Timeline')
plt.show()
```

```python
x = df['Yala Season']
y1 = df['Harvested 000 Ha.']
y2 = df['Average Yield Kg/Ha']

fig, ax1 = plt.subplots()

color = '#6ce5e6'
ax1.set_xlabel('Year', color='green')
ax1.set_ylabel('Value 1', color=color)
ax1.plot(x, y1, color=color, marker='o', linestyle='-', label='Line 1')
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x', labelcolor='green')

z1 = np.polyfit(x, y1, 1)
p1 = np.poly1d(z1)
ax1.plot(x, p1(x), "--", color=color, label="Trend Line for Line 1")

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Value 2', color=color)
# ax2.plot(x, y2, color=color, marker='o', linestyle='-', label='Line 2')
ax2.tick_params(axis='y', labelcolor=color)

z2 = np.polyfit(x, y2, 1)
p2 = np.poly1d(z2)
ax2.plot(x, p2(x), "--", color=color, label="Trend Line for Line 2")

fig.tight_layout()
legend = fig.legend(loc='upper left')
for text in legend.get_texts():
    text.set_color('green')

ax1.spines['top'].set_color('white')
ax1.spines['left'].set_color('white')
ax2.spines['bottom'].set_color('white')
ax2.spines['right'].set_color('white')

plt.gca().set_facecolor('none')
plt.title('Timeline', color='green')

plt.show()
```

```python
df['Uncultivated Land 000 Ha'] = df['Sown 000 Ha'] - df['Harvested 000 Ha.']

print(df)
```

```python
x = df['Yala Season']
y = df['Uncultivated Land 000 Ha']

plt.plot(x, y, marker='o', linestyle='-',label='Uncultivated Land')

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label="Trend Line")

plt.xlabel('Year')
plt.ylabel('Uncultivated Land 000 Ha')
plt.title('Timeline')

plt.tight_layout()
plt.legend(loc='upper left')

plt.grid(True)
plt.show()
```
